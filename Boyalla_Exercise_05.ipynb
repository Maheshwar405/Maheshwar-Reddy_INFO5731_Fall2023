```python
import nltk
nltk.download()
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords
import re
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def preprocess(sentence):
    sentence=str(sentence)
    sentence = sentence.lower()
    sentence=sentence.replace('{html}',"")
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url=re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num)
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
    stem_words=[stemmer.stem(w) for w in filtered_words]
    lemma_words=[lemmatizer.lemmatize(w) for w in stem_words]
    return " ".join(filtered_words)


dataset['cleanText']=dataset['Text'].map(lambda s:preprocess(s))
dataset.head()
```

```python
from gensim.models import Word2Vec
from sklearn.metrics import classification_report

# Tokenize the sentences
tokenized_text = [sentence.split() for sentence in dataset['cleanText']]
tokenized_text_test = [sentence.split() for sentence in dataset_test['cleanText']]

# Train Word2Vec model
word2vec_model = Word2Vec(sentences=tokenized_text, vector_size=100, window=5, min_count=1, workers=4)

# Function to calculate average word vectors for a sentence
def average_word_vectors(words, model, vocabulary, num_features):
    feature_vector = np.zeros((num_features,), dtype="float64")
    nwords = 0.

    for word in words:
        if word in vocabulary:
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, model.wv[word])

    if nwords:
        feature_vector = np.divide(feature_vector, nwords)

    return feature_vector

# Function to calculate average word vectors for all sentences
def averaged_word_vectorizer(corpus, model, num_features):
    vocabulary = set(model.wv.index_to_key)
    features = [average_word_vectors(sentence, model, vocabulary, num_features) for sentence in corpus]
    return np.array(features)

# Get feature vectors for training and validation sets
word2vec_train_features = averaged_word_vectorizer(tokenized_text, word2vec_model, 100)
word2vec_valid_features = averaged_word_vectorizer([sentence.split() for sentence in x_valid], word2vec_model, 100)
word2vec_test_features = averaged_word_vectorizer(tokenized_text_test, word2vec_model, 100)

# Train a classifier (e.g., MultinomialNB) on the Word2Vec features
classifier_word2vec = MultinomialNB()
model_word2vec = classifier_word2vec.fit(word2vec_train_features, y_train)

# Predictions on validation set
predictions_word2vec_validation = classifier_word2vec.predict(word2vec_valid_features)

# Evaluation metrics for validation set
print("Word2Vec Validation Metrics:")
print("Accuracy:", round(accuracy_score(y_valid, predictions_word2vec_validation) * 100, 2), "%")
print("Precision:", round(precision_score(y_valid, predictions_word2vec_validation, pos_label='0') * 100, 2), "%")
print("Recall:", round(recall_score(y_valid, predictions_word2vec_validation, pos_label='0') * 100, 2), "%")
print("F1 Score:", round(f1_score(y_valid, predictions_word2vec_validation, pos_label='0') * 100, 2), "%")

# Predictions on test set
predictions_word2vec_test = classifier_word2vec.predict(word2vec_test_features)

# Evaluation metrics for test set
print("Word2Vec Test Metrics:")
print("Accuracy:", round(accuracy_score(y_test, predictions_word2vec_test) * 100, 2), "%")
print("Precision:", round(precision_score(y_test, predictions_word2vec_test, pos_label='0') * 100, 2), "%")
print("Recall:", round(recall_score(y_test, predictions_word2vec_test, pos_label='0') * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, predictions_word2vec_test, pos_label='0') * 100, 2), "%")
```
